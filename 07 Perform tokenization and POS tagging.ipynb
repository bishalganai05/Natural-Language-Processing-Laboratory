```python
import nltk
import spacy
nlp = spacy.load("en_core_web_sm")
from nltk.tokenize import PunktSentenceTokenizer
from urllib import request
url="https://raw.githubusercontent.com/fbkarsdorp/python-course/master/data/austen-emma.txt"
response=request.urlopen(url)
text=response.read().decode('utf8')
Custom_sent_tokenizer=PunktSentenceTokenizer(text)
tokenize=Custom_sent_tokenizer.tokenize(text)
print("tokenized Sentences:\n")
print(tokenize)
print("\n PoS tagging: \n")
def process_content():
    for sentence in tokenize:
        tagged_words = nlp(sentence)
        for word in tagged_words:
            print(f"({word.text}, {word.pos_})")

process_content()
```
