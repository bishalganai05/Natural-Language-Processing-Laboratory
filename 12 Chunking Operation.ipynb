```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
from sklearn import tree
from nltk import *
text='A dog chased a cat'
tokens=word_tokenize(text)
print(tokens)
tag=nltk.pos_tag(tokens)
print(tag)
```

```python
grammer="NP:{<DT>?<JJ>*<NN.*>}"
cp=nltk.RegexpParser(grammer)
result=cp.parse(tag)
print(result)
print(result.pformat_latex_qtree())
result.pretty_print()
```
